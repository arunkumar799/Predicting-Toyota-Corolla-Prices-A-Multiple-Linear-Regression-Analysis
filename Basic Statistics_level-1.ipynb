## Standardization of Numerical Variables

### Explanation of Standardization

#### Standardization of Numerical Variables:This is especially important in machine learning and statistical modeling because it ensures that each feature contributes equally to the model, preventing any single feature from dominating due to its scale.

#### Standardization = X_bar -Mu/sigma

```python
New_volume=[]
for volume_Standardization in df['Volume']:
    Standardization1=(volume_Standardization-5.066667)/4.231602
    New_volume.append(Standardization1)
```

```python
New_AvgPrice=[]
for volume_Standardization in df['Avg Price']:
    Standardization2=(volume_Standardization-10453.433333)/18079.904840
    New_AvgPrice.append(Standardization2)
```

```python
New_Total_Sales_Value=[]
for volume_Standardization in df['Total Sales Value']:
    Standardization3=(volume_Standardization-33812.835556)/50535.074173
    New_Total_Sales_Value.append(Standardization3)
```

```python
New_Discount_Rate =[]
for volume_Standardization in df['Discount Rate (%)']:
    Standardization4=(volume_Standardization-15.155242)/4.220602
    New_Discount_Rate.append(Standardization4)
```

```python
New_Discount_Amount=[]
for volume_Standardization in df['Discount Amount']:
    Standardization5=(volume_Standardization-3346.499424)/4509.902963
    New_Discount_Amount.append(Standardization5)
```

```python
New_Net_Sales_Value=[]
for volume_Standardization in df['Net Sales Value']:
    Standardization6=(volume_Standardization-30466.336131)/46358.656624
    New_Net_Sales_Value.append(Standardization6)
```
